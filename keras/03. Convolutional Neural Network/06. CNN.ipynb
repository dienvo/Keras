## Data preprocessing

```python
x_train[0].shape
```

```python
#input image rwo and column

input_img_row=x_train[0].shape[0]
input_img_cols=x_train[0].shape[1]
```

```python
# reshape the input image to one dimension

x_train=x_train.reshape(x_train.shape[0],input_img_row,input_img_cols,1)
x_test=x_test.reshape(x_test.shape[0],input_img_row,input_img_cols,1)
```

```python
x_train[0]
```

60000 image 
28x28 pixels
1 dimension

```python
print(x_train.shape)
print(x_test.shape)
```

```python
x_train.ndim
```

```python
input_shape=(input_img_row,input_img_cols,1)
```

```python
input_shape
```

```python
# set all input image as a type float32

x_train=x_train.astype("float32")
x_test=x_test.astype("float32")
```

```python
x_train[0]
```

```python
# normalize the input data

x_train=x_train/255
x_test=x_test/255
```

```python
x_train[0]
```

```python
from keras.utils import np_utils
```

```python
# one hot encoder of the labels

y_train=np_utils.to_categorical(y_train)
y_test=np_utils.to_categorical(y_test)
```

```python
y_train[0]
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
num_classes=y_train.shape[1]
num_pixels=x_train.shape[1]*x_train.shape[2]
```

```python
print(num_classes)
print(num_pixels)
```

## Build the CNN layers

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten
from keras.layers import Conv2D,MaxPool2D
from keras.optimizers import SGD
```

```python
model=Sequential()
```

```python
model.add(Conv2D(filters=32,kernel_size=(3,3),activation="relu",input_shape=input_shape))
          
model.add(Conv2D(filters=64,kernel_size=(3,3),activation="relu"))
          
model.add(MaxPool2D(pool_size=(2, 2)))
          
model.add(Dropout(0.25))
          
model.add(Flatten())
          
model.add(Dense(units=128,activation="relu",))
          
model.add(Dropout(0.5))
          
model.add(Dense(units=num_classes,activation="softmax",))
```

## Compile the model

```python
model.compile(optimizer=SGD(0.01),
              loss="categorical_crossentropy",
              metrics=["accuracy"])
```

```python
model.summary()
```

## Train the model

```python
train=model.fit( x=x_train,
                 y=y_train,
                 batch_size=35,
                 epochs=10,
                 verbose = 1,
                 validation_data = (x_test, y_test))
```

## Evaluate the model

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

## Save the model

```python
model.save("cnn_model.h5")
```
